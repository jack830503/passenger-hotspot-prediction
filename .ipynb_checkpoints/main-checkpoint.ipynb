### Table: (zone, day, hour)

zone_day_hour_table

a baseline by building a table

last year's (zone, day, hour)

Score: 23/25 23.282899 2021/06/15 23:33:29 1

```python
import csv

train_hire_csv = "train_hire_stats.csv"
test_hire_csv = "test_hire_stats.csv"
submission_csv = "submission/submission-zone_day_hour_table.csv"

print("Build zone, day, hour table...")
zdh_table = {}  # zone day hour
with open(train_hire_csv, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:  # Zone_ID,Date,Hour_slot,Hire_count
        if row["Date"].split("-")[1] != "02":
            continue
        zdh = row["Zone_ID"], row["Date"].split("-")[-1], row["Hour_slot"]
        zdh_table[zdh] = row["Hire_count"]

print("Read test csv...")
with open(test_hire_csv, newline='') as csvfile:
    rows = [row for row in csv.DictReader(csvfile)]

print("Query table...")
for row in rows:  # Test_ID,Zone_ID,Date,Hour_slot,Hire_count
    zdh = row["Zone_ID"], row["Date"].split("-")[-1], row["Hour_slot"]
    row["Hire_count"] = zdh_table[zdh]

print("Write submission csv...")
with open(submission_csv, 'w', newline='') as csvfile:
    fieldnames = rows[0].keys()
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in rows:  # Test_ID,Zone_ID,Date,Hour_slot,Hire_count
        writer.writerow(row)

print("Done. Take your baseline submission.")
```

zone_weeknum1_weekday_hour_table

a baseline by building a table

last year's (zone, weeknum**+1**, weekday, hour)

Score: 22/25 14.789916 2021/06/16 04:51:45 3

```python
import csv

train_hire_csv = "train_hire_stats.csv"
test_hire_csv = "test_hire_stats.csv"
submission_csv = ("submission/"
    "submission-zone_weeknum1_weekday_hour_table.csv")

print("Build zone, weeknum, weekday, hour table...")
table = {}
with open(train_hire_csv, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:  # Zone_ID,Date,Hour_slot,Hire_count
        y, wn, wd = datetime.date.fromisoformat(row["Date"]).isocalendar()
        key = row["Zone_ID"], wn, wd, row["Hour_slot"]
        table[key] = row["Hire_count"]

print("Read test csv...")
with open(test_hire_csv, newline='') as csvfile:
    rows = [row for row in csv.DictReader(csvfile)]

print("Query table...")
for row in rows:  # Test_ID,Zone_ID,Date,Hour_slot,Hire_count
    y, wn, wd = datetime.date.fromisoformat(row["Date"]).isocalendar()
    key = row["Zone_ID"], wn + 1, wd, row["Hour_slot"]
    ## weeknum +1 to correlate the Chinese new year
    row["Hire_count"] = table[key]

print("Write submission csv...")
with open(submission_csv, 'w', newline='') as csvfile:
    fieldnames = rows[0].keys()
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in rows:  # Test_ID,Zone_ID,Date,Hour_slot,Hire_count
        writer.writerow(row)

print("Done. Take your baseline submission.")
```

### Table: avg (zone, workday, weekday, hour)

zone_workday_weekday_hour_table

a baseline by building a table

use last year's avg (zone, workday, weekday, hour),
(zone, weeknum+1, weekday, hour) if not found.

Score: 15/29 8.3637107 2021/06/16 22:22:26 4

```python
import csv
import datetime

train_hire_week_csv = "preprocess/train_hire_stats_week.csv"
test_hire_csv = "test_hire_stats.csv"
submission_csv = ("submission/"
    "submission-zone_workday_weekday_hour_table.csv")

def load_strs(s):
    strs = []
    for ln in s.split("\n"):
        strs.extend(ln.split())
    return strs

## from jummy1124/Taxi_Hotspot_prediction
holidays= load_strs("""
    2016-02-08 2016-02-09 2016-02-10 2016-02-11 2016-02-12 2016-02-29
    2016-04-04 2016-04-05 2016-06-09 2016-06-10 2016-09-15 2016-09-16
    2016-10-10 2017-01-02 2017-01-27 2017-01-30 2017-02-01 2017-02-27
    2017-02-28""")
makeupworkdays = "2016-06-04 2016-09-10 2017-02-18".split()

print("Build 2 tables...")
work_table = {}
week_table = {}
with open(train_hire_week_csv, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:  # Zone_ID,Date,Hour_slot,Hire_count
        hire = int(row["Hire_count"])
        work_key = tuple(
            row[i] for i in "Zone_ID Workday Weekday Hour_slot".split())
        if work_key in work_table:
            val, times = work_table[work_key]
            work_table[work_key] = (
                (val * times + hire) / (times+1), times+1 )
        else:
            work_table[work_key] = (hire, 1)
        week_key = tuple(
            row[i] for i in "Zone_ID Weeknum Weekday Hour_slot".split())
        week_table[week_key] = hire

print("Read test csv...")
with open(test_hire_csv, newline='') as csvfile:
    rows = [row for row in csv.DictReader(csvfile)]

print("Query table...")
for row in rows:  # Test_ID,Zone_ID,Date,Hour_slot,Hire_count
    y, wn, wd = datetime.date.fromisoformat(row["Date"]).isocalendar()
    workday = int(
        row["Date"] in makeupworkdays or (
            datetime.date.fromisoformat(row["Date"]).weekday() < 5 and \
            row["Date"] not in holidays
        )
    )
    work_key = row["Zone_ID"], str(workday), str(wd-1), row["Hour_slot"]
    week_key = row["Zone_ID"], str(wn+1), str(wd-1), row["Hour_slot"]
    if work_key in work_table:
        row["Hire_count"] = work_table[work_key][0]
    else:
        row["Hire_count"] = week_table[week_key]

print("Write submission csv...")
with open(submission_csv, 'w', newline='') as csvfile:
    fieldnames = rows[0].keys()
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in rows:  # Test_ID,Zone_ID,Date,Hour_slot,Hire_count
        writer.writerow(row)

print("Done. Take your baseline submission.")
```

## EDA

### Preprocess, plot dist

Generate columns "Year Month Day Weekday" from "Date".

Plot "Hire_count" distribution.

```python
import pickle
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train_hire_csv = "train_hire_stats.csv"
train_hire_week_csv = "preprocess/train_hire_stats_week.csv"
train_hire_week_pkl = "preprocess/train_hire_stats_week.pkl"

def load_strs(s):
    strs = []
    for ln in s.split("\n"):
        strs.extend(ln.split())
    return strs

## from jummy1124/Taxi_Hotspot_prediction
holidays= load_strs("""
    2016-02-08 2016-02-09 2016-02-10 2016-02-11 2016-02-12 2016-02-29
    2016-04-04 2016-04-05 2016-06-09 2016-06-10 2016-09-15 2016-09-16
    2016-10-10 2017-01-02 2017-01-27 2017-01-30 2017-02-01 2017-02-27
    2017-02-28""")
makeupworkdays = "2016-06-04 2016-09-10 2017-02-18".split()

df = pd.read_csv(train_hire_csv)
for i, s in enumerate("Year Month Day".split()):
    df[s] = df["Date"].apply(lambda s: s.split("-")[i])
df["Weeknum"] = df["Date"].apply(
    lambda s: datetime.date.fromisoformat(s).isocalendar()[1]
)
df["Weekday"] = df["Date"].apply(
    lambda s: datetime.date.fromisoformat(s).weekday()
)
df["Workday"] = df["Date"].apply(
    lambda s: int(
        s in makeupworkdays or (
            datetime.date.fromisoformat(s).weekday() < 5 and \
            s not in holidays
        )
    )
)

df = df["Zone_ID Date Year Month Day Weeknum Weekday Workday Hour_slot Hire_count".split()]
for i in df.columns:
    if i == "Date":
        df[i] = pd.to_datetime(df[i])
    else:
        df[i] = pd.to_numeric(df[i])
print(df)

df.to_csv(train_hire_week_csv, index=False)
with open(train_hire_week_pkl, "wb") as file:
    pickle.dump(df, file)

df["Hire_count"].plot.hist(bins=30)
plt.show()
print(df["Hire_count"].describe())
```

## Neural network methods

### Onehot MLP

#### Initial Settings

input is onehot encoded (zone, workday, weekday, hour)  
layers: dense 128 relu, dense 64 relu, dropout 0.1, dense 1  
optimizer: Adam lr=1e-3  
lr tuning: ReduceLROnPlateau

Why ReLU & Adam?  
Because the best practice on PPT page "Default DNN Configuration"  
(ELU + he_normal + Momentum SGD) not works.  
But this popular ReLU + Adam works quite well.

```python
import os
import glob
import pickle
import datetime
import numpy as np
import pandas as pd
import keras
from keras import layers as L
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

model_name = "zone_workday_weekday_hour_128_64_0.1_relu_center_again"
# model_name = "zone_workday_weekday_hour_feb"
model_save_dir = f"model/{model_name}/"
model_save_name = "model_{epoch:02d}_{loss:07.4f}_{val_loss:07.4f}"
csv_log_path = f"log/train.{model_name}.log.csv"
input_keys = "Zone_ID Workday Weekday Hour_slot".split()
# input_keys = "Zone_ID Weeknum Weekday Workday Hour_slot".split()

train_hire_week_pkl = "preprocess/train_hire_stats_week.pkl"
test_hire_csv = "test_hire_stats.csv"
submission_csv = f"submission/submission-{model_name}.csv"
load_model = None ##f"model/{model_name}"
predict_mcdropout = False
round_prediction = False

with open(train_hire_week_pkl, "rb") as file:
    df = pickle.load(file)

df = df[ df["Zone_ID"].isin([7,8,9,12,13,14,17,18,19]) ]
# df = df[ df["Month"] == 2 ]

enc = OneHotEncoder()
X = enc.fit_transform(df[input_keys]).toarray()
y = df["Hire_count"].values
print("Data shape:", X.shape, y.shape)
## center: (79056, 42)

X_train, X_test, y_train, y_test = train_test_split(X, y)
print("Train data shape:", X_train.shape, y_train.shape)
print("Test data shape:", X_test.shape, y_test.shape)

class MCDropout(keras.layers.Dropout):
    def call(self, inputs):
        return super().call(inputs, training=True)

##L.Dense(128, activation="elu", kernel_initializer="he_normal"),
model = keras.Sequential([
    L.Input(X.shape[-1]),
    L.Dense(128, activation="relu"),
    L.Dense(64, activation="relu"),
    L.Dropout(0.1),
    L.Dense(1)
], model_name)

model.compile(loss='mse',
    optimizer=keras.optimizers.Adam())#decay=1e-5))
    # keras.optimizers.SGD(momentum=0.9))

if load_model:
    model_path = max(glob.glob(os.path.join(load_model, "*")))
    print("Loading model:", model_path)
    model = keras.models.load_model(model_path)
else:
    history = model.fit(X, y, validation_data=(X_test, y_test),
        epochs=300,
        batch_size=1000,
        callbacks = [
        keras.callbacks.EarlyStopping(verbose=1, patience=15),
        # keras.callbacks.ReduceLROnPlateau(
        #     factor=0.5, patience=5, verbose=1, min_lr=5e-6),
        keras.callbacks.CSVLogger(csv_log_path, append=True),
        keras.callbacks.ModelCheckpoint(
            model_save_dir + model_save_name,
            verbose=1, save_best_only=True),
        keras.callbacks.TerminateOnNaN()
    ])

    # pd.DataFrame(history.history).plot(figsize=(8,5))
    # plt.grid(True)
    # plt.gca().set_ylim(0,1)
    # plt.show()

    input("Press ENTER to predict...")

def load_strs(s):
    strs = []
    for ln in s.split("\n"):
        strs.extend(ln.split())
    return strs

## from jummy1124/Taxi_Hotspot_prediction
holidays= load_strs("""
    2016-02-08 2016-02-09 2016-02-10 2016-02-11 2016-02-12 2016-02-29
    2016-04-04 2016-04-05 2016-06-09 2016-06-10 2016-09-15 2016-09-16
    2016-10-10 2017-01-02 2017-01-27 2017-01-30 2017-02-01 2017-02-27
    2017-02-28""")
makeupworkdays = "2016-06-04 2016-09-10 2017-02-18".split()

print("Prepare test data...")
df = pd.read_csv(test_hire_csv)
for i, s in enumerate("Year Month Day".split()):
    df[s] = df["Date"].apply(lambda s: s.split("-")[i])
df["Weeknum"] = df["Date"].apply(
    lambda s: datetime.date.fromisoformat(s).isocalendar()[1]
)
df["Weekday"] = df["Date"].apply(
    lambda s: datetime.date.fromisoformat(s).weekday()
)
df["Workday"] = df["Date"].apply(
    lambda s: int(
        s in makeupworkdays or (
            datetime.date.fromisoformat(s).weekday() < 5 and \
            s not in holidays
        )
    )
)

for i in df.columns:
    if i == "Date":
        df[i] = pd.to_datetime(df[i])
    else:
        df[i] = pd.to_numeric(df[i])
test_X = enc.transform(df[input_keys]).toarray()

print("Predict...")
if predict_mcdropout:
    pred = []
    for i in range(500):
        print(f"\r Predict {i+1}/500", end=" ")
        pred.append(model.predict(test_X))
    pred = np.stack(pred)
    print("\n", "Prediction shape:", pred.shape) # (500, 6048, 1)
    pred = pred.mean(axis=0)
    print("Mean shape:", pred.shape) # (6048, 1)
else:
    pred = model.predict(test_X)  # shape: (6048, 1)
if round_prediction:
    print("Round prediction to int")
    pred = np.rint(pred)

plt.figure(figsize=(20,5))
plt.plot(pred)
plt.show()

print("Save to csv...")
df["Hire_count"] = np.maximum(pred, 0)
df = df.drop("Year Month Day Weeknum Weekday Workday".split(), axis=1)
df.to_csv(submission_csv, index=False)

df
```
